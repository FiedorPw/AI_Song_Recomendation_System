```python
import os
import pickle

def map_ids(df, columns, mapping_dir='workspace_model_data/mappings'):
    """
    Mapuje unikalne wartości w zadanych kolumnach na unikalne indeksy zaczynając od 1
    i zapisuje mapowania do plików pickle.

    Args:
        df (pd.DataFrame): DataFrame zawierający kolumny do mapowania.
        columns (list): Lista nazw kolumn do mapowania.
        mapping_dir (str, optional): Ścieżka do katalogu, gdzie będą zapisywane mapowania. Defaults to 'workspace_model_data/mappings'.

    Returns:
        pd.DataFrame: DataFrame z nowymi kolumnami zawierającymi indeksy (np. 'user_id_idx').
    """
    os.makedirs(mapping_dir, exist_ok=True)
    for column in columns:
        unique_values = df[column].unique()
        id_to_idx = {id_: idx for idx, id_ in enumerate(unique_values, start=0)}
        idx_to_id = {idx: id_ for id_, idx in id_to_idx.items()}

        # Zapisz mapowanie id -> idx
        with open(os.path.join(mapping_dir, f'{column}_to_idx.pkl'), 'wb') as f:
            pickle.dump(id_to_idx, f)

        # Zapisz mapowanie idx -> id
        with open(os.path.join(mapping_dir, f'idx_to_{column}.pkl'), 'wb') as f:
            pickle.dump(idx_to_id, f)

        # Dodaj nową kolumnę z indeksami do DataFrame
        df[f'{column}_idx'] = df[column].map(id_to_idx)

    return df

```

```python
def load_mappings(columns, mapping_dir='workspace_model_data/mappings'):
    """
    Ładuje mapowania z plików pickle dla zadanych kolumn.

    Args:
        columns (list): Lista nazw kolumn, dla których chcesz załadować mapowania.
        mapping_dir (str, optional): Ścieżka do katalogu z mapowaniami. Defaults to 'workspace_model_data/mappings'.

    Returns:
        dict: Słownik zawierający mapowania 'column_to_idx'.
        dict: Słownik zawierający mapowania 'idx_to_column'.
    """
    column_to_idx = {}
    idx_to_column = {}

    for column in columns:
        with open(os.path.join(mapping_dir, f'{column}_to_idx.pkl'), 'rb') as f:
            column_to_idx[column] = pickle.load(f)
        with open(os.path.join(mapping_dir, f'idx_to_{column}.pkl'), 'rb') as f:
            idx_to_column[column] = pickle.load(f)

    return column_to_idx, idx_to_column
```

```python
event_map = {"skip": 0, "play": 1, "like": 1}


df["event_type"] = df["event_type"].map(event_map)
df["user_id"] = df["user_id"] - 101

columns_to_map = ['track_id']

# Mapowanie i dodanie kolumn indeksów
df = map_ids(df, columns_to_map)
```
